```python
from qiskit import *
from qiskit.visualization import plot_histogram

# Creamos tres qubits, dos clasicos y un circuito cuantico
q = QuantumRegister(2, name='q') # qubit cuantico
c = ClassicalRegister(2, name='c') # bit clasico
qc = QuantumCircuit(q, c)

# Aplicamos una compuerta Hadamard al primer qubit
qc.h(q[0])

# Aplicamos una compuerta controlada X entre el primer y segundo qubit
qc.cx(q[0], q[1])

# Aplicamos una compuerta Hadamard al primer qubit
qc.h(q[0])

# Medimos ambos qubits y almacenamos los resultados en los bits clásicos
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])

# Simulamos el circuito utilizando el simulador statevector_simulator
simulator = Aer.get_backend('statevector_simulator')
result = execute(qc, simulator, shots=1024).result()
counts = result.get_counts()

# Imprimimos los resultados de la simulación
print(counts)
plot_histogram(counts)
```

```python
from qiskit import *
from qiskit.visualization import plot_histogram

# Creamos tres qubits, dos clasicos y un circuito cuantico
q = QuantumRegister(2, name='q') # qubit cuantico
c = ClassicalRegister(2, name='c') # bit clasico
qc = QuantumCircuit(q, c)

# Aplicamos una compuerta Hadamard al primer qubit
qc.h(q[0])

# Aplicamos una compuerta controlada X entre el primer y segundo qubit
qc.cx(q[0], q[1])

# Aplicamos una compuerta Hadamard al primer qubit
qc.h(q[0])

# Medimos ambos qubits y almacenamos los resultados en los bits clásicos
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])

# Simulamos el circuito utilizando el simulador statevector_simulator
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc, simulator, shots=1024).result()
counts = result.get_counts()

# Imprimimos los resultados de la simulación
print(counts)
plot_histogram(counts)
```
